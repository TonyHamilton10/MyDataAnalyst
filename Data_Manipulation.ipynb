```python
data_jobs = pd.read_csv('C:/Users/user/Desktop/Machine Learning Full/data_jobs.csv')
```

```python
data_job_s = data_jobs.sample(frac = 0.5).copy()
```

```python
data_job_s.info()
```

```python
salary_year_median = data_job_s['salary_year_avg'].median()
salary_hrs_median = data_job_s['salary_hour_avg'].median()
```

```python
data_job_s['salary_year_avg'].replace(np.nan , salary_year_median , inplace = True)
```

```python
data_job_s['salary_hour_avg'].replace(np.nan ,salary_hrs_median , inplace = True )
```

```python
data_job_s.head()
```

```python
data_job_s['job_posted_date'] = pd.to_datetime(data_job_s['job_posted_date'])
```

```python
import ast
```

```python
data_job_s['job_skills'] = data_job_s['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
data_job_s.reset_index(inplace=True)
```

```python
data_job_s.drop('index' , axis = 1 , inplace = True)
```

```python
data_job_s.drop_duplicates(subset='job_title_short' , inplace=True)
```

```python
data_job_s = data_job_s.explode('job_skills')
```

```python
data_job_s.replace(to_replace = np.nan,value = None , method = 'bfill' , inplace = True)
```

```python
data_job_s.head()
```

```python
data_job_s['job_posted_month'] = data_job_s['job_posted_date'].dt.month
data_job_s['job_posted_year'] = data_job_s['job_posted_date'].dt.year
```

```python
data_job_s['job_posted_month_name'] = data_job_s['job_posted_date'].dt.strftime('%b')
```
