```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
%matplotlib inline
```

```python
superstore = pd.read_excel('C:/Users/user/Desktop/Machine Learning Full/data/archive (3)/SuperStore Sales DataSet.xlsx')
superstore.head()
```

```python
superstore.info()
```

```python
superstore['Order Date'] = pd.to_datetime(superstore['Order Date'])
superstore['Ship Date'] = pd.to_datetime(superstore['Ship Date'])
```

```python
superstore['Ship_Month'] = superstore['Ship Date'].dt.month
superstore['Ship_Year'] = superstore['Ship Date'].dt.year
superstore['Ship_Month_Name'] = superstore['Ship Date'].dt.strftime('%b')
```

```python
superstore['Ship_Month_Name'].unique()
```

```python
superstore.head()
```

```python
dep_piv = superstore.pivot_table(index = 'Ship_Month_Name' , columns='Category' , values='Profit' , aggfunc = 'sum')
dep_piv
```

```python
months = superstore[['Ship_Month' , 'Ship_Month_Name']].drop_duplicates()
months
```

```python
comb = pd.merge(dep_piv ,months , on='Ship_Month_Name' )
comb
```

```python
comb.sort_values('Ship_Month' , ascending=True , inplace=True)
```

```python
comb.drop('Ship_Month' , axis = 1 , inplace = True)
```

```python
comb.set_index('Ship_Month_Name' , inplace=True)
```

```python
sns.set_style('darkgrid')
comb.plot(kind = 'line')
plt.title('Profit Per Month')
plt.ylabel('Profit')
sns.despine()
```
