### Determining Rolling Stats

```python
rolmean= indexeddata.rolling(window=12).mean()

rolstd = indexeddata.rolling(window=12).std()

print(rolmean , rolstd)
```

```python
orig = plt.plot(indexeddata , color = 'blue' , label = 'Original')
mean = plt.plot(rolmean , color = 'red' , label = 'Rolling Mean')
std = plt.plot(rolstd , color = 'aquamarine' , label = 'Rolling Standard')
plt.legend(loc = 'best')
```

```python
from statsmodels.tsa.stattools import adfuller
print('Result of Dickey Filler Test')
dftest = adfuller(indexeddata['value'] , autolag='AIC')
```
